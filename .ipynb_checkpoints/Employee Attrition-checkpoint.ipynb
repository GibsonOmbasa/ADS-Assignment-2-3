```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
attrition = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
attrition.head()
```

```python
attrition.info()
```

```python
distance = attrition[["DistanceFromHome","JobRole", "Attrition"]]
distance.head()
```

```python
plt.style.use('ggplot')

distance.groupby(['JobRole', 'Attrition'])\
      .DistanceFromHome.count().unstack().plot.bar(legend=True)

plt.show()
```
